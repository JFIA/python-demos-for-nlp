```python
import pandas as pd
import numpy as np
import jieba
import codecs

def get_train_data(stopwords, path="sentiment_analysis/sentiment_analysis_trainingset.csv"):
    df = pd.read_csv(path, encoding="utf-8")
    # id = df["id"]
    content = df["content"]
    location_traffic_convenience = df["location_traffic_convenience"]
    content_list = []
    for text in content:
        token = jieba.cut(text)

        arr_temp = []
        for item in token:
            if (item not in stopwords) and (item not in '\t\n\r'):
                arr_temp.append(item)
        content_list.append(" ".join(arr_temp))
    
    return content_list
    
def loan_stopwords():
    stopwords = []
    with codecs.open('stopwords.txt', 'r', 'utf-8') as f:
        for each in f.readlines():
            each = each.strip('\n')
            each = each.strip('\r')
            stopwords.append(each)
    return stopwords

stopwords = loan_stopwords()

content = get_train_data(stopwords)
```

```python
import numpy as np

def build_sentence_vector(text, size,imdb_w2v):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in text:
        try:
            vec += imdb_w2v[word].reshape((1, size))
            count += 1.
        except KeyError:
            continue
    if count != 0:
        vec /= count
    return vec
train_vecs = np.concatenate([build_sentence_vector(z, 300,model) for z in content])

np.save('train_vecs.npy',train_vecs)
```
